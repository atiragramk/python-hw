```python
import seaborn as sns

penguins = sns.load_dataset("penguins")

penguins.head()
```

```python
#  displot

sns.kdeplot(data=penguins, x="body_mass_g")
```

```python
# regression
sns.regplot(y='bill_depth_mm', x='body_mass_g', data=penguins)
```

```python
# scatterplot

sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=penguins)
```

```python
# lineplot
sns.lineplot(x="sex", y="flipper_length_mm", data=penguins)
```

```python
#  Box Plot
sns.boxplot(x="species", y="flipper_length_mm", data=penguins)
```

```python
# Bar Plot
sns.barplot(x="species", y="flipper_length_mm", data=penguins)
```

```python
# Count Plot
sns.countplot(x="species", data=penguins)
```

```python
# swarmplot
sns.catplot(y="body_mass_g", x="sex", kind="swarm", data=penguins)
```
